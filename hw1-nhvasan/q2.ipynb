```python
from sklearn import datasets
import pandas as pd
import numpy as np
import seaborn as sns
import nbconvert
```

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris['data'], columns = iris.feature_names)
iris_df['species'] = iris['target']

# index 0: setosa
# index 1: versicolor
# index 2: virginica
iris_df['species'] = np.where(iris_df['species'] == 0, 'setosa',
                     np.where(iris_df['species'] == 1, 'versicolor',
                     np.where(iris_df['species'] == 2, 'virginica', 'NA')))
iris_df
```

### b) Plots
#### Comparative Boxplots

```python
sepal_length_df = iris_df[['sepal length (cm)', 'species']]
sepal_length_df.boxplot(by='species')
```

```python
sepal_width_df = iris_df[['sepal width (cm)', 'species']]
sepal_width_df.boxplot(by='species')
```

```python
petal_length_df = iris_df[['petal length (cm)', 'species']]
petal_length_df.boxplot(by='species')
```

```python
petal_width_df = iris_df[['petal width (cm)', 'species']]
petal_width_df.boxplot(by='species')
```

#### Petal and Sepal Distributions

**Petal**

```python
petal_data = iris_df[['petal length (cm)', 'petal width (cm)', 'species']]
petal_data
```

```python
sns.scatterplot(data=petal_data, x='petal length (cm)', y='petal width (cm)', hue='species')
```

**Sepal**

```python
sepal_data = iris_df[['sepal length (cm)', 'sepal width (cm)', 'species']]
sepal_data
```

```python
sns.scatterplot(data=sepal_data, x='sepal length (cm)', y='sepal width (cm)', hue='species')
```

**Overview (cm)**

*Setosa:*
- sepal length: lowest sepal length (mean ~5.0)
- sepal width: highest sepal width (mean ~3.4)
- petal length: lowest petal length (mean ~1.5)
- petal width: lowest petal width (mean ~0.25)

*Versicolor:*
- sepal length: medium sepal length (mean ~5.8)
- sepal width: lowest sepal width (mean ~2.8)
- petal length: medium petal length (mean ~4.5)
- petal width: medium petal width (mean ~ 1.25)

*Virginica:*
- sepal length: largest sepal length (mean ~6.5)
- sepal width: medium sepal width (mean ~3.0)
- petal length: highest petal length (mean ~5.5)
- petal width: highest petal width (mean ~2.0)

**Summary (Rules)**
* If the sepal length is small (~5.0cm), the sepal width is large (~3.4cm), but the petal length and width are small (~1.5cm and 0.25cm respectively), then the flower is an Iris Setosa.
* If the sepal length is medium (~5.8cm), the sepal width the small (~2.8cm), but the petal length and width are medium-sized (~4.5cm and ~1.25cm respectively), then the flower is an Iris Versicolor.
* If the sepal length is large (~6.5cm), the sepal width is medium-sized (~3.0cm), and the petal length and width are also large (~5.5cm and ~2.0cm respectively), then the flower is an Iris Virginica.
